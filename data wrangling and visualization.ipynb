```python
def user(course,C1,C2,C3,C4,C5):
    if course == "AIC":
        return np.array([C1,C2,C3,C4,C5]).sum()
    elif course == "IOT":
        return np.array([C1,C2,C3,C4,C5]).prod()
    elif course =="CNC":
        return sum(np.array([C1,C2,C3,C5,C4]).cumsum())
```
