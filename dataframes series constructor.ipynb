```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# possible data inputs to series constructor

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series((-1,-2,-3,-4,-5))
```

```python
print(s1)
print(s2)
```

```python
s1 = pd.Series([1,2,3,4,5], index=['A','B','C','D','E'])
s2 = pd.Series((-1,-2,-3,-4,-5), index=['i','ii','iii','iv','v'], name='Loop')
s2.index.name = "Roman Index"
print(s1,s2, sep='\n\n')
```

```python
dic = {'a':1,'b':2,'c':3,'d':4}
s3 = pd.Series(dic)
print(s3)
```
